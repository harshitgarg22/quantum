# Determine if function is constant or balanced

This notebook is based on Jonathan Hui's [medium article](https://medium.com/@jonathan_hui/qc-quantum-algorithm-with-an-example-cf22c0b1ec31).

A function is called **constant** if it outputs the same number {0, 1} irrespective of the input, or the number of times it is run.

A function is called **balanced** if it outputs one of {0, 1} half the time, irrespective of the input.
